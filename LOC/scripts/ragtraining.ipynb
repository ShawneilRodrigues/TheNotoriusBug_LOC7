```python
import pandas as pd

all_nodes = []

for csv_file in csv_files: 
    df = pd.read_csv(csv_file)
    # Filter out rows with NaN values in 'bio'
    df = df.dropna(subset=['bio'])
    nodes = []

    for _, row in df.iterrows():
        # Extract fields
        bio = row.get("bio", "N/A")  # Used for similarity search
        emp_id = row.get("Employee ID", "N/A")
        name = row.get("Name", "N/A")
        role = row.get("Role", "N/A")
        phone = row.get("Phone Number", "N/A")
        email = row.get("Email", "N/A")
        gender = row.get("Gender", "N/A")
        address = row.get("Address", "N/A")

        # Combine all columns into full context text for output
        full_context = (
            f"Employee ID: {emp_id}\n"
            f"Name: {name}\n"
            f"Role: {role}\n"
            f"Phone: {phone}\n"
            f"Email: {email}\n"
            f"Gender: {gender}\n"
            f"Address: {address}\n"
            f"Bio: {bio}"
        )

        # Create a TextNode with 'bio' as the text for similarity search
        # and full context as metadata for final output
        node = TextNode(
            text=bio,  # Used for similarity search
            metadata={
                "Full Context": full_context,  # Retain all columns for output
                "Employee ID": emp_id,
                "Name": name,
                "Role": role,
                "Phone": phone,
                "Email": email,
                "Gender": gender,
                "Address": address
            }
        )
        nodes.append(node)

    # Print the count of created TextNodes for the current CSV
    print(f"Created {len(nodes)} TextNodes for file {csv_file}.")  # File-specific output

    # Add nodes to the global list
    all_nodes.extend(nodes)  # Append nodes to the global list

print(f"Total nodes added to the vector store: {len(all_nodes)}")
```


# Alert Data Processing and Vector Store Setup

## Data Processing Pipeline

1. **Initial Setup**
    - Installing required packages for LlamaIndex, HuggingFace embeddings, and ObjectBox
    - Setting up necessary imports and configurations

2. **LLM Model Configuration**
    - Using Llama-3.1-8B-Instruct model
    - Configured with temperature 0.1 and context window of 130K tokens
    - GPU optimization enabled for inference

3. **Data Ingestion**
    - Processing multiple CSV files containing alert data
    - Files contain alert codes, categories, and related information
    - Each CSV contains different alert categories (2.2.x, 2.3.x, 3.1.x)

4. **Node Creation**
    - Converting CSV rows into TextNodes for vector storage
    - Each node contains:
      - Primary text from alert body
      - Metadata including alert code, category, severity, and explanations
    - Filtering out rows with missing body content

5. **Vector Store Implementation**
    - Using ObjectBox as vector store
    - HuggingFace embeddings (BAAI/bge-small-en model)
    - 384-dimensional embeddings with cosine similarity
    - Persistent storage for future retrieval

6. **Query System**
    - Similarity search capability with top-k retrieval
    - Full context preservation in search results
    - Support for natural language queries


```python
# Initialize the ObjectBox vector store
from llama_index.core import StorageContext, VectorStoreIndex, Settings
embed_model = HuggingFaceEmbedding(model_name="BAAI/bge-small-en")

vector_store = ObjectBoxVectorStore(
    embedding_dimensions=384,  
    distance_type=VectorDistanceType.COSINE,   
    db_directory="objectbox_data_25_12_24_body",  
    clear_db=True,             
    do_log=True               
)

Settings.embed_model = embed_model

# Create a storage context using ObjectBox
storage_context = StorageContext.from_defaults(vector_store=vector_store)

print("ObjectBox vector store initialized.")
index = VectorStoreIndex(nodes=all_nodes, storage_context=storage_context)
index.storage_context.persist()
```

```python
# Create a retriever from the index
retriever = index.as_retriever(similarity_top_k=3)

# Example query text
query_text = "software interns"

# Get similar nodes
nodes = retriever.retrieve(query_text)

# Print the results
for node in nodes:
    print("Text:", node.text)
    print("Metadata:", node.metadata)
    print("-" * 80)
```

```python
from llama_index.core import load_index_from_storage, StorageContext

# Load the existing storage context
storage_context = StorageContext.from_defaults(
	vector_store=ObjectBoxVectorStore(
		embedding_dimensions=384,
		distance_type=VectorDistanceType.COSINE,
		db_directory="objectbox_data_25_12_24_body",
		clear_db=False,
		do_log=True
	)
)

# Load the persisted index
index = VectorStoreIndex.from_vector_store(
	vector_store=storage_context.vector_store,
	embed_model=embed_model
)

print("Index loaded successfully.")
```

```python

from llama_index.llms.ollama import Ollama

llm = Ollama(
    model="llama3.1",
    temperature=0.1,
    context_window=1000,
    request_timeout=480 # Increase timeout to 120 seconds
)
def get_context_and_query_llm(user_message, retriever=retriever, llm=llm):
    try:
        # Get relevant context using the retriever
        retrieved_nodes = retriever.retrieve(user_message)
        
        # Extract relevant contexts and their metadata
        contexts = []
        for node in retrieved_nodes:
            contexts.append(node.metadata["Full Context"])
        
        # Combine contexts
        combined_context = "\n\n".join(contexts)
        
        # Create prompt with context
        prompt = f"""
You are an assistant tasked with classifying customer feedback using the provided context.

### Customer Feedback
"{user_message}"

### Context (from Knowledge Base or RAG System)
The following information is retrieved from the context and defines valid values for categories, subcategories, risk levels, and alert codes:
{combined_context}

### Task
Based on the provided context and feedback, predict the following:
1. Category (select ONLY from the categories in the context; if no risk, return 'None').
2. Subcategory (select ONLY from the subcategories in the context; if no risk, return 'None').
3. Risk Level (select ONLY from the levels defined in the context; if no risk, return 'None').
4. Alert Code (select ONLY from the codes defined in the context; if no risk, return '0.0.0').

### Additional Instructions
1. Highlight **up to 2-3 sentences or phrases** in the Customer Feedback message that are most critical to your decision-making process. Do not highlight more than 3 sentences. (None if Alert Code is 0.0.0).
2. Focus only on the **most important** sentences or phrases that directly contributed to your decision.
3. DO NOT create or assume values that are not explicitly present in the context.

### Rules
1. If the Alert Code is "0.0.0", then all other fields (Category, Subcategory, Risk Level) MUST be "None".
2. If any of the fields (Category, Subcategory, Risk Level) are not "None", then the Alert Code MUST NOT be "0.0.0".
3. DO NOT create or assume values that are not explicitly present in the context.
4. Use ONLY the data provided in the context to make your predictions.

### Output Format
Output your answer ONLY in JSON format like this:
{{
    "Category": "<value>",
    "Subcategory": "<value>",
    "Risk Level": "<value>",
    "Alert Code": "<value>",
    "Highlighted Sentences": ["<sentence1>", "<sentence2>", "<sentence3>"]
}}

DO NOT include any additional notes or explanations. Only return valid JSON.
"""
        
        # Query the LLM
        response = llm.complete(prompt)
        
        return response.text
    except Exception as e:
        print(f"Error occurred: {str(e)}")
        return "Sorry, I encountered an error while processing your request. Please try again."
    return response.text


user_msg = "We have identified errors in the discounts applied to your invoices, leading to undercharging. This has caused confusion and financial discrepancies in our accounts. The incorrect discounts have also led to discrepancies in our financial records, making it difficult for us to maintain accurate accounts. We have had to involve our finance team to investigate the discrepancies, which has resulted in additional costs and delays. We request an immediate correction and clarification on the correct discounts. Additionally, we expect a detailed explanation of the steps you will take to ensure accurate discount application in the future."
response = get_context_and_query_llm(user_msg)
print(response)
```
